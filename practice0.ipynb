```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
```

```python
uni = pd.read_csv('Universities.csv')
uni.describe()
uni.head()
```

```python
uni.data = uni.iloc[:,1:]  #only numerical values
uni.data.head()
```

```python
UNI = uni.data.values  #converting to array
UNI
```

```python
uni_normal = scale(UNI)
uni_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(uni_normal)
pca_values
```

```python
pca = PCA(n_components = 6)
Pca_values = pca.fit_transform(uni_normal)
Pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals= 4)*100)
var1
```

```python
plt.plot(var1,color = "red")
```

```python
x = pca_values[:,0:1]
y= pca_values[:,1:2]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']),uni['Univ']],axis=1)
```

```python
finalDf
```

```python
# prompt: plot scatter plot

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
sns.scatterplot(data = finalDf, x = 'pc1', y = 'pc2', hue = 'Univ')
```

```python
import pandas as pd
data = pd.read_csv("NewspaperData (1).csv")
```

```python
data[['daily', 'sunday']].corr()
```

```python
import seaborn as sns
sns.distplot(data['daily'])
```

```python
sns.distplot(data['sunday'])
```

```python
import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data).fit()
```

```python
model.params
```

```python
model.rsquared
```

```python
model.tvalues
```

```python
 model.pvalues
```
